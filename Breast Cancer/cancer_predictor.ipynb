# 3)Handling Null Values

# 5)Handling Categorical Values

```python
cancer.head()
```

```python
# from the dataset it is clearly evident that we have only one column of categorical data --- diagnosis---significant
#--- perform categorical handling next
```

```python
# two ways to handle categorical data---- 1) label encoding 2) get_dummies method
# using label encoding
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

cancer["diagnosis"] = le.fit_transform(cancer["diagnosis"]) # replacing the transformed values back into original data
```

```python
cancer["diagnosis"].unique() 

# encoded values---1 /0
#1----M
#0----B (Alphaetical order is considered for assigning)
```

# 6) Split into Dependent and Independent Variables: x,y

```python
cancer.head(2)
```

```python
x = cancer.iloc[:,1:31].values
y= cancer.iloc[:,0:1].values
```

```python
x.shape
```

```python
y.shape
```

```python
x
```

```python
y
```

# 7)Split into train and test

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

# 8)Feature Scaling

```python
# to remove any outliers present in the data

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
 x_test
```

# 9)Model Building

```python
from sklearn.linear_model import LogisticRegression
logregtrain = LogisticRegression()

logregtrain.fit(x_train,y_train)
```

# 10)Testing

```python
logregtest = logregtrain.predict(x_test)
```
